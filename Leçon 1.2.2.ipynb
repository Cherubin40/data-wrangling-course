```python
import numpy as np
import pandas as pd

source = "data/HDR_2007-2008_Table_22..xls"

df = pd.read_excel(source)
df.head(12)
```

```python
df = pd.read_excel(source, header=9) # Positione l'en-tête à la ligne 7
df = df.dropna(how="all", axis=1)    # Supprime toute les colonnes vides "na"
df = df.dropna(how="all", axis=0).reset_index(drop=True) # Recommence le décompte des les ligne
df.head(6)
```

```python
# Pour écrir les en-têtes sur plusieurs lignes sur une seule ligne
columns = []

for c in df.columns:
    columns.append([str(v) for v in df[c][:4] if pd.notnull(v)])

columns
```

```python
columns = [
    'HDI rank',
    'Country',
    'Electricity consumption per capita (kilowatt-hours) 2004',
    'Reference 1',
    'Electricity consumption per capita (% change) 1990-2004',
    'Reference 2',
    'Electrification rate (%) 2000-05a',
    'Reference 3',
    'Population without  electricity (millions) 2005',
    'Reference 4',
    'GDP per unit of energy use (2000 PPP US$ per kg of oil equivalent) 2004',
    'Reference 5',
    'GDP per unit of energy use (% change) 1990-2004',
    'Forest area % of total land area (%) 2005',
    'Reference 6',
    'Forest area Total (thousand sq km) 2005',
    'Reference 7',
    'Forest area Total change (thousand sq km) 1990-2005',
    'Reference 8',
    'Forest area Average annual change (%) 1990-2005'
]

df.columns = columns

df = df.drop(np.arange(4), axis=0).reset_index(drop=True)
df.head()
```

```python
df.tail(30)
```

```python
df = df.drop(np.arange(180,len(df)), axis=0).reset_index(drop=True)
```

```python
df.tail()
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
# Se débarrasser des rangées avec les catégories HDI
# https://stackoverflow.com/a/27360130 & https://stackoverflow.com/a/12098586
df = df.drop(df[df["HDI rank"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)
# Référence pour np.where https://docs.scipy.org/doc/numpy/reference/generated/numpy.where.html
# Pensez que cela ressemble à une déclaration du type "if" : where(test est vrai, vrai-quelque chose, faux-quelque chose)
# ceci peut aussi être imbriqué : where(test is true, true-thing, where(new-test is true, true-thing, false-thing))
df.loc[:, "HDI category"] = np.where(df["HDI rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI rank"] > 70,
                                              "MEDIUM HUMAN DEVELOPMENT",
                                              "HIGH HUMAN DEVELOPMENT"
                                             )
                                    )
df.sort_values(by="HDI rank", inplace=True)
df.head()
```

```python
destination = "data/world-urban-population-data.csv"
df.to_csv(destination, index=False)
```
